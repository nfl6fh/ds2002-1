```python
import sys
import re
```

```python
def extract_names(filename):
  """
  Given a file name for baby.html, returns a list starting with the year string
  followed by the name-rank strings in alphabetical order.
  ['2006', 'Aaliyah 91', Aaron 57', 'Abagail 895', ' ...]
  """
  # +++your code here+++
  with open(filename, 'r') as f:
    result = []
    text = f.read()
    year = re.search(r'Popularity in ([0-9]{4})', text)
    if not year:
      sys.stderr.write('Couldn\'t find the year!')
    year_proper = year[0][-4:]
    result.append(year_proper)
    name_lines = re.findall(r'<td>[0-9]+</td><td>[A-Z][a-z]+</td><td>[A-Z][a-z]+</td>', text)
    # name_list = []
    for line in name_lines:
      rank = re.search(r'[0-9]+', line)[0]
      names = re.findall(r'[A-Z][a-z]+', line)
      for name in names:
        result.append(f'{name} {rank}')
    result.sort()
  return result
```
